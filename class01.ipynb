```python
import chardet

def check_encoding(filename):
    rawdata = open(filename, 'rb').read()
    result = chardet.detect(rawdata)
    encoding = result['encoding']
    confidence = result['confidence']
    return encoding, confidence

file_path = '000001.csv'# 添加文件路径
encoding, confidence = check_encoding(file_path)
print(f"Encoding: {encoding}, Confidence: {confidence}")

# 本单元格是导入数据的代码
data = pd.read_csv('000001.csv') # 这一句是导入CSV文件的命令
data
print(data) # 让我看看数据是什么样子
```

```python
type(data)
print(data.columns)
print(data.columns.values)

print(data[['Day']])
import pandas as pd

# 假设您已经加载了数据到data变量中
# 如果没有，请先加载数据，例如：
# data = pd.read_csv('your_file.csv')

# 创建带样式的表格（修复语法错误）
styled = data.style\
    .set_table_styles([
        {'selector': 'table',
         'props': [('border', '2px solid red'),
                   ('border-collapse', 'collapse')]},
        {'selector': 'th, td',
         'props': [('border', '1px solid red'),
                   ('background-color', 'yellow'),
                   ('padding', '8px'),
                   ('text-align', 'left')]},
        {'selector': 'th',
         'props': [('background-color', "#ffcc00"),  # 稍深一点的黄色
                   ('font-weight', 'bold')]}
    ])

# 将样式表格导出为HTML
html_output = styled.to_html()

# 保存HTML文件（修复语法错误）
with open('styled_table.html', 'w', encoding='utf-8') as f:
    f.write(html_output)

print("已生成带样式的HTML表格文件: styled_table.html")
print("请在浏览器中打开此文件查看效果")
```
